```python
class Simple_Linear_Regression :
    # Attributes  if any 
    def __init__(self,X_train,X_test,y_train,y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
    def Mean(self, population):
        self.population = population
        return np.sum(self.population) / len(self.population)
    def Variance(self, population):
        self.population = population
        return np.sum((self.population - self.Mean(self.population)) ** 2) / len(self.population)
    def Covariance(self):
        return np.sum((self.X_train - self.Mean(self.X_train)) * (self.y_train - self.Mean(self.y_train))) / len(self.X_train)
    def Coefficents(self):
        self.beta1 = self.Covariance() / self.Variance(self.X_train)
        self.beta0 = self.Mean(self.y_train) - self.beta1 * self.Mean(self.X_train)
        return self.beta0, self.beta1
    def simple_linear_regression(self):
        return self.Coefficents()[1] * self.X_test + self.Coefficents()[0]
    def MSE(self):
        return np.sum((self.y_test - self.simple_linear_regression()) **2 ) / len(self.y_test)
    def Evaluate(self):
        return self.MSE()
    def Vizualize(self):
        plt.figure(figsize=(10,8))
        plt.title('PREDICTION VS ACTUAL ', fontsize=24 , fontstyle='italic')
        plt.scatter(self.X_test ,self.y_test , alpha =0.5, label='Y')
        plt.plot(self.X_test, self.simple_linear_regression(), linewidth=2.0)
        return plt.show         
    # You can make this on your own by plotting  Points and Line 
                        # plt.scatter () , plt.plot()
         
```
